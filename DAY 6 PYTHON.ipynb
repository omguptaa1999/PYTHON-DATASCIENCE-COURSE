WORKING WITH LARGE AMOUNT OF DATA

```python
for pk in pd.read_csv('modified.csv', chunksize = 5):
    print("CHUNK PK")
    print(pk)
```

```python
new_pk = pd.DataFrame(columns = pk.columns)
for pk in pd.read_csv('modified.csv', chunksize = 5):
    results = pk.groupby(['Type 1']).count()
    new_pk = pd.concat([new_pk, results])
    print(new_pk)
```
